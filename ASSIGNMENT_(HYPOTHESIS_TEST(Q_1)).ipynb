```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Cutlets.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
unitA=pd.Series(data.iloc[:,0])
unitA
```

AS PER GIVEN QST
H0:u1=u2
H1: u1 =/ u2

```python
unitB=pd.Series(data.iloc[:,1])
unitB
```

```python
plt.hist(unitA)
plt.title('unitA')
plt.show()
```

```python
plt.hist(unitB)
plt.title('unitB')
plt.show()
```

```python
#its a two tailed test
p_value=stats.ttest_ind(unitA,unitB)
p_value
```

```python
p_value[1]
```

```python
#compare p value with significance value=0.05
# pvalue is greater than significance value
# we fail reject ho
# there is no significance difference
```
